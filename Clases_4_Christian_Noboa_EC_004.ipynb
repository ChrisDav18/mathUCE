#Imagen y Pre imagen  üîÑ
Se aplican a conjuntos y nos arrojan conjuntos:

**Imagen**:conjunto de n√∫meros reales que son imagen por de los elementos de su dominio.

**Preimagen:** Es un conjunto de puntos en $A$ cuya imagen pertenece a ese conjunto.

Supongamos que tenemos la funci√≥n:

$f: A$-> $B$, x-> $f(x)$

- Sea $E‚äÜA$, $f(E)$:{$f(x)œµB:xœµE$}, significa que $f(E)$ es un conjunto de las imagenes cuyos elemetos de $E$ existen en $B$, es decir la imagen de E.
- Sea $F‚äÜB$, $f^-1(F):=${$xœµA:f(x)œµF$}, esto seria la preimagen de F (osea que podemos usar la funci√≥n inversa para encontrar la preimagen).

Ejemplos:

$f(A)‚äÜB$

$f: R$->$R, X$ -> $f(x)=x^2+1$

$f(R)$={$f(x)œµR:xœµR$}, = {$x^2+1:xœµR$}

$f(R)$={$xœµR: x‚â•1$} o $[1,+‚àû[‚äÜR$

La preimagen seria: $f^-1(]-10,0[)$ y la definici√≥n seria: {$xœµR:x^2+1œµ]-10,0[$}

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 1

x = np.linspace(-10, 10, 400)

y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x) = x^2 + 1$', color='blue')
plt.title('Gr√°fica de $f(x) = x^2 + 1$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

#Ejercicios: üìù

¬øCu√°l seria la preimagen del conjunto $f^-1:([0,2])$?

{$xœµR: x^2+1œµ[0,2]$} Como buscamos la preimagen debemos analizarlo algebraicamente.

= {$xœµR: 0‚â§x^2+1‚â§2$}

= {$xœµR:-1‚â§x^2‚â§1$}  y buscamos que x son igual o menores a 1

= {$xœµR:-1‚â§x‚â§1$} finalmente podemos decir que la preimagen de  $f^-1:([0,2])$ es $[-1,1]$

Gr√°ficamente se ve:

```python
import numpy as np
import matplotlib.pyplot as plt

def f_inv(y):
    return np.sqrt(y - 1)

y = np.linspace(1, 2, 400)

x_positive = f_inv(y)
x_negative = -f_inv(y)

plt.figure(figsize=(8, 6))
plt.plot(x_positive, y, label=r'$f^{-1}(y) = \sqrt{y - 1}$', color='red')
plt.plot(x_negative, y, label=r'$f^{-1}(y) = -\sqrt{y - 1}$', color='blue')
plt.title('Gr√°fica de la funci√≥n inversa $f^{-1}(y)$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

Ejemplo:

$g: R$ -> $R, x$->$g(x)=sin(x)$

¬øCu√°l es la imagen?

$g(R)= [-1,1]$  todos los puntos de los reales tienen una imagen que pertenece al conjunto $[-1,1]$

Seg√∫n la definici√≥n:

$g(R)$={$g(x)œµR:xœµR$}   
={$sin(x):xœµR$}

La funcion de seno siempre tiene un conjunto del $[-1,1]$

Gr√°ficamente:

```python
import numpy as np
import matplotlib.pyplot as plt

def g(x):
    return np.sin(x)

x = np.linspace(-2 * np.pi, 2 * np.pi, 400)

y = g(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$g(x) = \sin(x)$', color='purple')

plt.scatter([-np.pi/2, np.pi/2], [-1, 1], color='red', zorder=5)
plt.text(-np.pi/2, -1.1, '(-œÄ/2, -1)', color='red', horizontalalignment='center')
plt.text(np.pi/2, 1.1, '(œÄ/2, 1)', color='red', horizontalalignment='center')

plt.title('Gr√°fica de $g(x) = \sin(x)$')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
