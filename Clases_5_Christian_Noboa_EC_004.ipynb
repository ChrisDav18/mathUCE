# Funci√≥n de utilidad en Econom√≠a üìà
En econom√≠a, una funci√≥n de utilidad es una herramienta utilizada para expresar el nivel de satisfacci√≥n o bienestar que una persona obtiene al consumir bienes. Esta funci√≥n calcula la utilidad o satisfacci√≥n obtenida a partir de la cantidad de bienes consumidos.
- **Funci√≥n de Utilidad con un solo Bien**

Pensemos que solo tenemos diponible el bien del aguaüíß, la funci√≥n de utilidad se le puede representar como (u(x)), (x) es la cantidad de agua que se consumio y (u(x)) es la utilidad que se obtiene. Se lo puede denotar como: $[Agua(x)$-> $Utilidad(u(x))]$

```python
import matplotlib.pyplot as plt
import numpy as np

def utilidad_agua(x):
    return np.sqrt(x)

x = np.linspace(0, 10, 100)
y = utilidad_agua(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Utilidad de Consumo de Agua')
plt.xlabel('Cantidad de Agua (Litros)')
plt.ylabel('Utilidad')
plt.title('Funci√≥n de Utilidad para Consumo de Agua')
plt.grid(True)
plt.legend()
plt.show()
```

Podemos visualizar esta conexi√≥n usando una curva que ilustra c√≥mo la utilidad crece al incrementar el consumo de agua. No obstante, la utilidad marginal disminuye gradualmente, lo que indica que cada adici√≥n de agua proporciona menos satisfacci√≥n que la anterior. Por ejemplo, el primer vaso de agua puede ofrecer mucho placer en un d√≠a caluroso, pero los vasos adicionales, como el tercero o cuarto, proporcionan menos satisfacci√≥n en comparaci√≥n.

- Funci√≥n de Utilidad con dos bienes üìà

Pongamos en situaci√≥n en que no solo tenemos ahora aguaüíß sino que tambien tenemos panü•ñ, Esta funci√≥n de utilidad ahora dependera de la cantidad consumida de ambas, es decir, $[u(x,y)]$

$x$ representa la cantidad de agua e $y$ la cantidad de pan. La utilidad se representa en un gr√°fico tridimensional, donde los ejes $x$ e $y$ corresponden a las cantidades de los dos bienes, y el eje vertical representa la utilidad. Esta representaci√≥n gr√°fica muestra una superficie de utilidad.

Podemos imaginar esta superficie como una forma de concha, donde cada punto en la superficie representa un nivel espec√≠fico de utilidad asociado con una combinaci√≥n particular de $x$ e $y$. Si observamos desde el plano $xy$, veremos curvas parecidas a las del caso unidimensional.

Ejemplos:
- (log(x))
-(‚àöx)

Graficamente:


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 10, 400)

y = np.log(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='log(x)', color='b', linewidth=2)

plt.title('Gr√°fica de log(x)')
plt.xlabel('x')
plt.ylabel('log(x)')
plt.grid(True)
plt.legend()

plt.show()

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 400)

y = np.sqrt(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sqrt(x)', color='g', linewidth=2)

plt.title('Gr√°fica de sqrt(x)')
plt.xlabel('x')
plt.ylabel('sqrt(x)')
plt.grid(True)
plt.legend()

plt.show()


```

Estas funciones son crecientes pero cada vez tienen una pendiente menor, lo que muestra es el **principio de utilidad marginal decreciente**.

#Notaci√≥n Matem√°tica de la Funci√≥n de Utilidad üìä
- Definici√≥n de la funci√≥n de utilidad: [$U:R+^2$->$R+]$, Significa que la funci√≥n de utilidad $U$ toma como entrada a un par de n√∫meros reales o negativos $((R+^2))$ y se produce un n√∫mero real no negativo $((R+))$.

- Expresi√≥n de la funci√≥n de utilidad: $[(x,y)$->$u(x,y)]$, Significa que $((x,y))$ son las cantidades de los 2 bienes y el nivel de utilidad de esas cantidades es: $(u(x.y))$
- Conjuntos de Nivel: $[${$(x,y)‚ààR+^2:u(x,y)=10$}] , significa que $((x,y))$ son el conjunto de todos los pares en $(R^2)$ y dan una utilidad de 10.


```python
import numpy as np
import matplotlib.pyplot as plt

def u(x, y):
    return np.log(x) + np.sqrt(y)

x = np.linspace(0.1, 5, 400)
y = np.linspace(0, 10, 400)
X, Y = np.meshgrid(x, y)
Z = u(X, Y)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='plasma', edgecolor='none', alpha=0.8)

contours = ax.contour(X, Y, Z, levels=[1, 2, 3, 4, 5, 10], cmap='inferno', linestyles="solid")

ax.set_xlabel('Bien 1')
ax.set_ylabel('Bien 2')
ax.set_zlabel('Utilidad U(x, y)')
ax.set_title('Funci√≥n de Utilidad y Conjuntos de Nivel')

plt.show()
```

Estos conjuntos de nivel se los representa por curvas de nivel conocidos como $(isocuantas)$

#Notaci√≥n Matem√°ticaüìä
- Conjunto de nivel:

[$f^-1$({$10$})={$(x,y)‚ààR^2:u(x,y)‚àà${$10$}}]

[{$(x,y):‚àöxy=10$}], al simplicarlo queda: [{$(x,y):xy=100$}]

- Generalizaci√≥n para un nivel de utilidad:
[$f^-1$({$u0$})={$(x,y)‚ààR+^2:y=uo^2/x$}], significa que para un nivel de utilidad $(u0)$, se mantienne la relaci√≥n $(x)$ e $(y)$ en la ecuaci√≥n de antes.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def u(x, y):
    return np.sqrt(x * y)

x = np.linspace(0.1, 10, 400)
y = np.linspace(0.1, 10, 400)
X, Y = np.meshgrid(x, y)
Z = u(X, Y)

u_levels = [5, 10, 15, 20, 25]

fig = plt.figure(figsize=(14, 9))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='plasma', edgecolor='none', alpha=0.7)

colors = ['blue', 'green', 'red', 'purple', 'orange']
for i, u0 in enumerate(u_levels):
    x_contour = np.linspace(0.1, 10, 400)
    y_contour = (u0**2) / x_contour
    z_contour = u(x_contour, y_contour)
    ax.plot(x_contour, y_contour, z_contour, label=f'u0 = {u0}', linewidth=2, color=colors[i])

ax.set_xlabel('Bien 1')
ax.set_ylabel('Bien 2')
ax.set_zlabel('Utilidad U(x, y)')
ax.set_title('Superficie de la Funci√≥n de Utilidad y Conjuntos de Nivel')

ax.legend()
plt.show()

import numpy as np
import matplotlib.pyplot as plt

def u(x, y):
    return np.sqrt(x * y)

def preimage(u0, x):
    return u0**2 / x

u_levels = [5, 10, 15, 20, 25]

x = np.linspace(0.1, 10, 400)

colors = ['blue', 'green', 'red', 'purple', 'orange']

plt.figure(figsize=(10, 6))

for i, u0 in enumerate(u_levels):
    y = preimage(u0, x)
    plt.plot(x, y, label=f'u0 = {u0}', color=colors[i])

plt.xlabel('Bien 1')
plt.ylabel('Bien 2')
plt.title('Conjuntos de Nivel de la Funci√≥n de Utilidad')
plt.legend()

plt.grid(True)
plt.show()


```

#Funci√≥n Sobreyectiva, Biyectiva, Inyectiva, Inversa y Compuesta üîÑ




- **Funci√≥n sobreyectiva** üéØ

Es sobreyectiva cuando su imagen cubre todo el conjunto $(B):[f(A)=B]$, lo que significa que $(y‚ààB)$ tiene que existir al menos $(x‚ààA)$ lo que daria como resultado: $[f(x)=y]$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

x = np.linspace(-5, 5, 100)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = x^2', color='blue')
plt.title('Gr√°fica de una Funci√≥n Sobreyectiva')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

- **Funci√≥n Inyectiva**üîç

Es inyectiva cuando se le asigna elementos distintos de (A) a los elementos distintos de (B), es decir: $[‚àÄx,x'‚ààA, x‚â†x'$ -> $f(x)‚â†f(x')]$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definici√≥n de la funci√≥n inyectiva
def f(x):
    return x

# Rango de valores para x
x = np.linspace(-5, 5, 100)
y = f(x)

# Creaci√≥n del gr√°fico
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = x', color='blue')
plt.title('Gr√°fica de una Funci√≥n Inyectiva')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

- **Funci√≥n Biyectiva**üîó

Es biyectiva cuando es tanto **Inyectiva** como **Sobreyectiva**. Denotado como: $[f $es biyectiva$‚áîf $$es inyectiva ‚àß f es sobreyectiva$]

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3

x = np.linspace(-2, 2, 100)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = x^3', color='blue')
plt.title('Funci√≥n Biyectiva f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.show()

```

- **Funci√≥n inversa** üîÅ

Toda funci√≥n biyectiva tiene su funci√≥n inversa $(f^-1)$ y va desde $(B)$ hasta $(A)$.

Se denota como: $(f^-1)$

 Funci√≥n inversa (g) cumple con: $[g:B$ ->$A][g(y)=x  $tal que $ f(x)=y]$

Tiene que cumplir 2 propiedades:
- [‚àÄx‚ààA, g(f(x))=x]
- [‚àÄx‚ààB, f(g(y))=y]


#Funciones CompuestasüîÄ
**Definici√≥n**üìú
- Partimos de 3 conjuntos: $E$,$F$ y $G$
- Definimos 2 funciones: (f) y (g):

$(f:E$->$F)$, donde $(f(x)=f(x))$

$(g: F$ -> $G)$, donde $(g(y)=g(y))$

**La funci√≥n compuesta** $(g $ o $f)$ es definidad con la notaci√≥n: $[(g$ o $f)(x))=g(f(x))]$, lo que significa que se toma un elemento $x$ de $E$, se le aplica la funci√≥n $f$ para tener  $(f(x))$ en $F$, despues aplicar la funci√≥n $g$ a $(f(x))$ para obtener $(g(f(x)))$ en $(G)$.

**Definici√≥n de Funcionesüìä:**

-Funci√≥n $(f)$ de los reales a un intervalo: $[f:R$->$[1,+‚àû[]$ $[f(x)=x^2+1]$

-Funci√≥n $(g)$ del intervalo $([1,+‚àû])$ a los reales positivos: $[g:[1,+‚àû[$-> $R^+] [g(y)=‚àöy]$

**Composici√≥n de $(g)$y$(f)$üîÑ:**

Se la define como: $[g o f: R$-> $R^+]$= $[(g/circf)(x)$

$=g(f(x))=g(x^2+1) = ‚àöx^2+1] rt{x^2+1}$


**Ejemplo de Funci√≥n compuestaüåü:**

$(h(x)):[h(x)=‚àöx^2+1]$


```python
import numpy as np
import matplotlib.pyplot as plt

def h(x):
    return np.sqrt(x**2 + 1)

x = np.linspace(-5, 5, 400)
y = h(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='h(x) = sqrt(x^2 + 1)', color='blue')
plt.title('Gr√°fica de la funci√≥n h(x)=g(f(x))')
plt.xlabel('x')
plt.ylabel('h(x)')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.show()
```
