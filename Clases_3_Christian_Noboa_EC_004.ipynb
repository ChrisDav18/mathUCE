# **¿Qué es una computadora y como sirven los programas?**💻

Una computadora puede ejecutar operaciones básicas, como suma, resta, multiplicación y división. La persona siempre tiene que activarle a la computadora mediante un software, sin embargo, si programamos una computadora podemos hacerla que sea más veloz que un humano.

**Código**💻: Es el lenguaje que utilizamos para comunicarnos con la máquina.

**Ejemplo de lo que podemos programar:**🚗

Supongamos que quieres calcular la velocidad media de un viaje en coche. Tienes la distancia recorrida y el tiempo que te ha llevado. ¿Cómo harías que una computadora calcule la velocidad media?

Tendrías que darle instrucciones muy claras y paso a paso:

- Acepta un número que represente la distancia recorrida
- Acepta un número que represente el tiempo que ha llevado el viaje
- Divide el primer número (distancia) por el segundo número (tiempo) para obtener la velocidad media
- Muestra el resultado de una forma que sea fácil de entender para un humano.

```python

distancia = float(input("Introduce la distancia recorrida en kilómetros: "))


tiempo = float(input("Introduce el tiempo del viaje en horas: "))


velocidad_media = distancia / tiempo


print(f"La velocidad media del viaje es de {velocidad_media:.2f} kilómetros por hora.")
```

#Lenguaje máquina vs Lenguaje de alto nivel🤖🚀
**Lenguaje máquina**🖥️

Conocido tambien como **IL** (Instruction List), es el alfabeto de las computadoras compuesto por el conjunto de símbolos simples y primarios que podemos usar para dar comandos a una computadora.

**Lenguajes de alto nivel**🌉

Las computadoras y los humanos necesitan un lenguaje común, algo parecido a un "puente" que conecten estos 2 mundos. Para ello usamos lo que se conoce como **lenguaje de alto nivel**, estos utilizan símbolos, palabras y convenciones legibles para humanos, lo que a su vez, permite expresar comandos más complejos a las computadoras.

**Codigo fuente**💻

Se refiere a un programa escrito en un lenguaje de programación de alto nivel. Tiene todas las instrucciones para que la computadora ejecute un programa. El archivo que contiene ese código se lo conoce como **archivo fuente**.

Gracias a estos lenguajes, se puede traducir el formato para que las computadoras puedan entender y ejecutar.

#Compilación vs Interpretación ⚙️

La programación informática es la acción de componer los elementos del lenguaje de programación seleccionando el orden en el que se efectuara.

Se debe cumplir ciertos requisitos:
- **Alfabeticamente**🔠
- **Léxicamente**📖
- **Sintácticamente**🧾
- **Semánticamente**🧠

Ahora entra la compilación y la interpretación.

Existen 2 maneras para tranformar un programa de un lenguaje de alto nivel a un lenguaje que la computadora pueda entender:

- **Compilación**📦: Un código fuente se traduce una vez a lenguaje de máquina, se crea un archivo ejecutable que es distribuido y ejecutado en cualquier computadora sin una necesidad del código fuente original.
- **Interpretación**🕮: Un código fuente se traduce una vez que se ejecuta el programa, el intérprete lee y ejecuta el código línea por línea, por lo que se necesita tanto un código fuente como el intérprete para ejecutar el programa.

#¿Qué hace el intérprete?🤔
**Código fuente📄: archivo de texto**✍️
Al escribir un programa se crea un **archivo de computadora**, este programa de computadora es en realidad un fragmento de texto, por lo que el código fuente generalmente se coloca en **archivos de texto**.

**Nota**: Tiene que ser **texto puro** sin ningún tipo de fuentes, colores, imágenes u otros medios.

**El intérprete**🖥️

Al tener un archivo de código fuente, se debe invocar al interprete que podrá leer tu archivo.

El intérprete lee el código fuente de arriba a abajo y de izquierda a derecha.

Los pasos que realiza para verificar tu código son:
- Alfabéticamente🔠
- Léxicamente📖
- Sintácticamente🧾
- Semánticamente🧠

Si se encuentra un error, saldrá un **mensaje de error**⚠️

Cada línea debe ser ejectuada por separada por lo que el trío **"Leer- Verificar - Ejecutar"** se repite varias veces y algunas partes del código pueden ejecutarse varias veces.

#Sistemas Numéricos -continuación- y Cardinales🔢✨

Hay otras 3 formas para construir conjuntos más grandes:
- **Conjunto sencillo**🧩 {1,2,3} = $Ω$
- **Set potencial o set de poder de omega**⚡️ Constituido por todos los subconjuntos que pueden formarse a partir de elementos de omega:  $P(Ω)=${$∅$,{1},{2}, {3}, {1,2}, {1,3}, {2,3}, $Ω$}, tambien puede existir el subconjunto vacio: { } $= ∅$

La cantidad de elementos cuando trabajamos en conjuntos se le conoce como el **cardinal** y para el ejemplo de arriba, se denota como: $card(Ω)=3$ y el $card(P(Ω))=2^{\text{card}(\Omega)}$.

- **Producto cartesiano**🔄 Se concibe como: Sea $a,b$ conj.  
AxB={$(a,b): aϵA,bϵB$}          
Ejemplo de como se construye:                                          
A={$1,2,3$}              
B={$0,1$}            
AxB={$(1,0)(1,1)(2,0)(2,1)(3,0)(3,1)$}      
Aqui tenemos 6 elementos y podemos sacar la relación de cardinales.📊       
$card(AXB)=card(A)*card(B)$  -> en el caso que los conjuntos sean finitos.     
Es muy útil cuando trabajamos con conjuntos más interesantes como por ejemplo $N*N$={(0,0),(0,1),(0,2)...} o tambien (1,0),(1,1),(1,2), etc. este es un conjunto bastante grande. 🌌             
Otro ejemplo:                             
$R*R$={$(x,y):xϵR,yϵR$} que esto da origen a: $R^n$={$(x1,x2,...xn): x1ϵR, x2ϵR,...xnϵR$}.
Otra forma de denotarlo es: $R^n=${$(x1,x2,...xn):∀iϵ[1,n],x,ϵR$}.

##División de los conjuntos

- **Conjunto finito:** Son aquellos que tienen un número finito de elementos, como el afabeto.🔤
- **Conjunto infinito ♾️:** Son aquellos que tienen un número infinito de elementos.

Nota: Existen diferentes tamaños de infinitos.

En términos de cardinales podemos establecer una relación de infinitos como $card(N)=card(Z)$, esto debido a que podemos establecer una regla de conteo que relacione ambos infinitos. Pero no siempre podremos relacionar diferentes infinitos.

**Dentro de los conjuntos infinitos tenemos:**

-Los números Naturales y enteros relativos son ejemplos de **infinitos contables o numerables.**

-Los reales y más grandes de ellos son **infinitos no contables o no numerables.**
